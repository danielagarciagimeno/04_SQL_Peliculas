```python
import pandas as pd

# Cargo los enlaces en un diccionario por si luego queremos ampliar o reducir el origen de datos
files = {
    'Amazon Prime': '/Users/monzon8/Documents/Ironhack/PROYECTOS/04_SQL/titles_Amazon_Prime.csv',
    'AppleTV': '/Users/monzon8/Documents/Ironhack/PROYECTOS/04_SQL/titles_AppleTV.csv',
    'Disney': '/Users/monzon8/Documents/Ironhack/PROYECTOS/04_SQL/titles_Disney.csv',
    'HBO': '/Users/monzon8/Documents/Ironhack/PROYECTOS/04_SQL/titles_HBO.csv',
    'Netflix': '/Users/monzon8/Documents/Ironhack/PROYECTOS/04_SQL/titles_Netflix .csv',
    'Paramount': '/Users/monzon8/Documents/Ironhack/PROYECTOS/04_SQL/titles_Paramount.csv'
}
```

```python
# Cargo cada archivo y añado la columna platform 
dataframes = []
for platform, file_path in files.items():
    df = pd.read_csv(file_path)
    df['platform'] = platform
    dataframes.append(df)
```

```python
columns_info = {platform: df.columns.tolist() for platform, df in zip(files.keys(), dataframes)}
columns_info
```

```python
#Hago un concat de todos lso DF teniendo como referencia el 'id' y agrego una columna con las múltiples plataformas siempre que tenga la misma 'id'
combinado_df = pd.concat(dataframes, ignore_index=True)
combinado_df = combinado_df.groupby('id', as_index=False).agg({
    **{col: 'first' for col in combinado_df.columns if col != 'platform'},
    'platform': lambda x: ', '.join(sorted(set(x)))  # Combine platform names for duplicate ids
})
```

```python
combinado_df['seasons'] = combinado_df['seasons'].fillna(0)
```

```python
combinado_df
```

```python
combinado_df = combinado_df.rename(columns={'id': 'id_titles'})
combinado_df = combinado_df.rename(columns={'genres': 'genre_title'})
combinado_df = combinado_df.rename(columns={'production_countries': 'country_title'})
combinado_df = combinado_df.rename(columns={'platform': 'platform_title'})

```
